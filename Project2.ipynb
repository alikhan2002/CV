```python
import cv2
import numpy as np
        
cap=cv2.VideoCapture(0)

width=int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height=int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

cv2.namedWindow('Script1')

while True:
    ret,frame=cap.read()
    
    frame[0:height//2,0:width//2]=cv2.blur(frame[0:height//2,0:width//2],ksize=(100,100))
    frame[0:height//2,width//2:width]=cv2.blur(frame[0:height//2,width//2:width],ksize=(10,10))
    frame[height//2:height,0:width//2]=cv2.blur(frame[height//2:height,0:width//2],ksize=(50,50))
    frame[height//2:height,width//2:width]=cv2.blur(frame[height//2:height,width//2:width],ksize=(30,30))
    
    cv2.imshow('Script1',frame)

    if cv2.waitKey(1) & 0xFF==ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
        
cap=cv2.VideoCapture(0)

width=int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height=int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

cv2.namedWindow('Script2')

while True:
    ret,frame=cap.read()
    
    #frame=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    
    top_left=cv2.resize(frame,(0,0),fx=0.5,fy=0.5)
    top_right=cv2.resize(frame,(0,0),fx=0.5,fy=0.5)
    bottom_left=cv2.resize(frame,(0,0),fx=0.5,fy=0.5)
    bottom_right=cv2.resize(frame,(0,0),fx=0.5,fy=0.5)
    top_left=cv2.blur(top_left,ksize=(100,100))
    top_right=cv2.blur(top_right,ksize=(10,10))
    bottom_left=cv2.blur(bottom_left,ksize=(50,50))
    bottom_right=cv2.blur(bottom_right,ksize=(30,30))
    
    frame[0:height//2,0:width//2]=top_left
    frame[0:height//2,width//2:width]=top_right
    frame[height//2:height,0:width//2]=bottom_left
    frame[height//2:height,width//2:width]=bottom_right
    print(top_left.shape, end="sfs\n")
    print(frame.shape)
    cv2.imshow('Script2',frame)     

    if cv2.waitKey(1) & 0xFF==ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2

pt1=[0,0]
pt2=[0,0]
pt1_draw=[0,0]
pt2_draw=[0,0]
drawing=False
move=False
check=0
list1=[]
list2=[]
list3=[]
list4=[]
listX1,listX2=[],[]
X1,X2=50,50
is2=0
listCheck=[]

def Applyfilter(event,x,y,flags,param):
    global pt1,pt2,pt1_draw,pt2_draw,drawing,move,check,list1,list2,list3,list4,listX1,listX2,X1,X2,is2,listCheck
    
    if event==cv2.EVENT_LBUTTONDOWN:
        if drawing==False:
            pt1=[x,y]
            drawing=True
            
    if event==cv2.EVENT_MOUSEMOVE:
        if drawing==True:
            pt2=[x,y]
            if pt1[0]<pt2[0] and pt1[1]<pt2[1]:
                pt1_draw=pt1
                pt2_draw=pt2
            elif pt1[0]<pt2[0] and pt1[1]>pt2[1]:
                pt1_draw[0]=pt1[0]
                pt2_draw[0]=pt2[0]
                pt1_draw[1]=pt2[1]
                pt2_draw[1]=pt1[1]
            elif pt1[0]>pt2[0] and pt1[1]<pt2[1]:
                pt1_draw[0]=pt2[0]
                pt2_draw[0]=pt1[0]
                pt1_draw[1]=pt1[1]
                pt2_draw[1]=pt2[1]
            elif pt1[0]>pt2[0] and pt1[1]>pt2[1]:
                pt1_draw=pt2
                pt2_draw=pt1
            move=True
            
    if event==cv2.EVENT_LBUTTONUP:
        if drawing==True:
            if is2%2==0:
                listCheck.append(0)
            else:
                listCheck.append(1)
            list1.append(pt1_draw[0])
            list2.append(pt1_draw[1])
            list3.append(pt2_draw[0])
            list4.append(pt2_draw[1])
            listX1.append(X1)
            listX2.append(X2)
            pt1=[0,0]
            pt2=[0,0]
            pt1_draw=[0,0]
            pt2_draw=[0,0]
            X1=50
            X2=50
            drawing=False
            move=False
        
cap=cv2.VideoCapture(0)

cv2.namedWindow('Script3')

cv2.setMouseCallback('Script3',Applyfilter)

while True:
    ret,frame=cap.read()
    
    if drawing==True and move==True:
        if pt1_draw[0]==pt2_draw[0] or pt1_draw[1]==pt2_draw[1]:
            check=1
        if check==0:
            if is2%2==0:
                frame[pt1_draw[1]:pt2_draw[1],pt1_draw[0]:pt2_draw[0]]=cv2.blur(frame[pt1_draw[1]:pt2_draw[1],pt1_draw[0]:pt2_draw[0]],ksize=(X1,X2))
            else:
                hsv=cv2.cvtColor(frame[pt1_draw[1]:pt2_draw[1],pt1_draw[0]:pt2_draw[0]],cv2.COLOR_BGR2HSV)
                hsv[:,:,2]=cv2.equalizeHist(hsv[:,:,2])
                hsv=cv2.cvtColor(hsv,cv2.COLOR_HSV2BGR)
                frame[pt1_draw[1]:pt2_draw[1],pt1_draw[0]:pt2_draw[0]]=hsv
        check=0
    
    for i in range(len(list1)):
        if listCheck[i]==0:
            frame[list2[i]:list4[i],list1[i]:list3[i]]=cv2.blur(frame[list2[i]:list4[i],list1[i]:list3[i]],ksize=(listX1[i],listX2[i]))
        else:
            hsv=cv2.cvtColor(frame[list2[i]:list4[i],list1[i]:list3[i]],cv2.COLOR_BGR2HSV)
            hsv[:,:,2]=cv2.equalizeHist(hsv[:,:,2])
            hsv=cv2.cvtColor(hsv,cv2.COLOR_HSV2BGR)
            frame[list2[i]:list4[i],list1[i]:list3[i]]=hsv

    
    cv2.imshow('Script3',frame)
  
    if cv2.waitKey(1) & 0xFF==ord('q'):
        break
    elif cv2.waitKey(1) & 0xFF==ord('a'):
        X1+=2
        X2+=2
    elif cv2.waitKey(1) & 0xFF==ord('s'):
        if X1>2:
            X1-=2
            X2-=2
    elif cv2.waitKey(1) & 0xFF==ord('d'):
        if len(list1)!=0:
            list1.pop()
            list2.pop()
            list3.pop()
            list4.pop()
            listX1.pop()
            listX2.pop()
    elif cv2.waitKey(1) & 0xFF==ord('f'):
        is2+=1

cap.release()
cv2.destroyAllWindows()
```
