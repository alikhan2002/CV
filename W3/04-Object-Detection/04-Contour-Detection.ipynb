# Contour Detection

## External vs Internal Contours

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('../DATA/internal_external.png',0)
```

```python
img.shape
```

```python
plt.imshow(img, cmap='gray')
```

### findContours

 cv::RETR_EXTERNAL = 0,
  cv::RETR_LIST = 1,
  cv::RETR_CCOMP = 2,
  cv::RETR_TREE = 3,
  cv::RETR_FLOODFILL = 4

```python
image, contours, hierarchy = cv2.findContours(img, cv2.RETR_CCOMP,cv2.CHAIN_APPROX_SIMPLE)
```

```python
type(contours)
```

```python
len(contours)
```

```python
type(hierarchy)
```

```python
hierarchy
```

```python
hierarchy[0][0][3]
```

```python
# Draw Ecternal Contour

external_contour = np.zeros(image.shape)

for i in range(len(contours)):
    
    if hierarchy[0][i][3] == -1:
        
        cv2.drawContours(external_contour,contours, i, 255,-1)
```

```python
plt.imshow(external_contour,cmap='gray')
```

```python
image_internal = np.zeros(image.shape)

for i in range(len(contours)):
    
    if hierarchy[0][i][3] != -1:
        
        cv2.drawContours(image_internal,contours, i, 255,-1)
```

```python
plt.imshow(image_internal,cmap='gray')
```
