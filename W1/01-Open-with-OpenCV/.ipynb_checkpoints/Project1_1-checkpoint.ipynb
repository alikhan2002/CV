# Drawing with mouse

# Рисовалка

```python
import cv2 
import numpy as np
import math
def nothing(x):
    pass

    
img = np.zeros((1024, 1024, 3), np.uint8)

drawing = False

ix, iy = -1, -1
img = np.zeros((1024, 1024, 3), np.uint8)

def draw_rectangle(event, x, y, flags, param):
    global ix, iy, drawing
    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True
        ix, iy = x, y
        
    elif event == cv2.EVENT_MOUSEMOVE:
        
        if drawing == True:
            cv2.rectangle(img, (ix, iy), (x, y), (b,g,r), -1)
    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
        cv2.rectangle(img, (ix, iy), (x, y), (b,g,r), -1)

def draw_line(event, x,y, flags, param):
    global ix, iy, drawing
    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True
        ix, iy = x, y

    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing == True:
            cv2.line(img, pt1=(ix,iy), pt2=(x, y), color=(b,g,r), thickness=s)
    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
        cv2.line(img, pt1=(ix,iy), pt2=(x, y), color=(b,g,r), thickness=s)
    

        
def draw_circle(event, x,y, flags, param):
    global ix, iy, drawing
    cv2.imshow('image', img)
    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True
        ix, iy = x, y

    elif event == cv2.EVENT_MOUSEMOVE:

        if drawing == True:
            cv2.circle(img,center=(ix,iy), radius=int(math.sqrt((x-ix)**2 + (y-iy)**2)), color=(b,g,r), thickness=-1)
    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
        cv2.circle(img,center=(ix,iy), radius=int(math.sqrt((x-ix)**2 + (y-iy)**2)), color=(b,g,r), thickness=-1)
        
def draw_real(event, x,y, flags, param):
    global ix, iy, drawing
    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True
        ix, iy = x, y

    elif event == cv2.EVENT_MOUSEMOVE:

        if drawing == True:
            cv2.circle(img,center=(x,y), radius=20, color=(b,g,r), thickness=-1)
    elif event == cv2.EVENT_LBUTTONUP:
        cv2.circle(img,center=(ix,iy), radius=20, color=(b,g,r), thickness=-1)
        drawing = False

cv2.namedWindow('image', cv2.WINDOW_NORMAL)

# Create trackbars for color change
cv2.createTrackbar('R','image',0,255,nothing)
cv2.createTrackbar('G','image',0,255,nothing)
cv2.createTrackbar('B','image',0,255,nothing)
cv2.createTrackbar('Thickness', 'image',1,5,nothing)
cv2.setMouseCallback('image', draw_line)

while True:
    
    cv2.imshow('image', img)
    
    if cv2.waitKey(20) & 0xFF == 27: # Esc Bottom
        break
    if cv2.waitKey(1) & 0xFF == ord('1'): 
        cv2.setMouseCallback('image', draw_line)
    elif cv2.waitKey(1) & 0xFF == ord('2'): 
        cv2.setMouseCallback('image', draw_rectangle)
    elif cv2.waitKey(1) & 0xFF == ord('3'): 
        cv2.setMouseCallback('image', draw_circle)
    elif cv2.waitKey(1) & 0xFF == ord('4'): 
        cv2.setMouseCallback('image', draw_real)
    
    r = cv2.getTrackbarPos('R','image')
    g = cv2.getTrackbarPos('G','image')
    b = cv2.getTrackbarPos('B','image')
    s = cv2.getTrackbarPos('Thickness','image')
cv2.destroyAllWindows()



```
