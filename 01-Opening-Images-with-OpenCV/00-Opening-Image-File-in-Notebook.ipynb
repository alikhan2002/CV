```python
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

# Resize Image

```python
img_rgb.shape
```

```python
img = cv2.resize(img_rgb,(1300,250))
```

```python
plt.imshow(img)
```

# By ratio

```python
w_ratio = 0.2
h_ratio = 0.2
```

```python
new_img = cv2.resize(img_rgb,(0,0),img,w_ratio,h_ratio)
plt.imshow(new_img)
```

# Flipping Images

```python
# Along central x axis
new_img = cv2.flip(new_img,0)
plt.imshow(new_img)
```

```python
# Along central y axis
new_img = cv2.flip(new_img,1)
plt.imshow(new_img)
```

```python
# Along both axis
new_img = cv2.flip(new_img,-1)
plt.imshow(new_img)
```

# Saving Image Files
